```python
data = pd.read_csv('drive/MyDrive/ML-datasets-master/kc_house_data.csv', index_col=None)
```

```python
new_data = data.drop(['id', 'date', 'zipcode', 'sqft_living15', 'sqft_lot15', 'lat', 'long', 'sqft_lot', 'waterfront', 'floors', 'sqft_above', 'view', 'yr_renovated'], axis=1)
new_data
```

```python
x = new_data['sqft_basement']
is_basement = []

for i in x:
  if i>0:
    is_basement.append(1)
  else:
    is_basement.append(0) 

y = np.reshape(is_basement, (-1,1))
new_data['is_basement'] = y
new_data = new_data.drop(['sqft_basement'], axis = 1)
```

```python
corr = new_data.corr()
sns.heatmap(corr, annot=True)
```
